```python
%matplotlib notebook

import pandas as pd
```

```python
df = pd.read_csv('CityPairs.csv')
df
```

```python
df = df.loc[df['AustralianPort'].isin(['Melbourne', 'Sydney', 'Brisbane', 'Perth'])]
df
```

```python
df = df.filter(items=[
    'AustralianPort',
    'Year',
    'Passengers_Total',
    'Freight_Total_(tonnes)',
    'Mail_Total_(tonnes)'
])
df
```

```python
grouped_year_port = df.groupby(['AustralianPort', 'Year'])
df = grouped_year_port.sum()
df
```

```python
df.plot.scatter(x='Passengers_Total', y='Freight_Total_(tonnes)', s=df['Mail_Total_(tonnes)'])
```
